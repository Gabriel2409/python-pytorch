```python
import torch
from torch import nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
```

```python
F.selu
```

```python
nb_features = 100
nb_ex = 10000
x = torch.empty(nb_ex,nb_features).normal_(mean=0,std=1)
w = torch.empty(nb_features).normal_(mean=0,std=1).unsqueeze(0)
z = F.selu(x @ w.T)
_ = plt.hist(z.numpy(), bins=100)
```
